```python
import json
from pprint import pprint

with open('./flex/flex_data.json') as json_data:
    accounts = json.load(json_data)
    json_data.close()
    pprint(accounts)
```

```python
def validate_account_totals(account):
    declared_value = float(account["value"])
    validation_log = {}
    if "items" in account:
        values = []
        for i in account["items"]:
            values.append(float(i["value"]))
        derived_value = round(sum(values), 2)
        if derived_value == declared_value:
           status = "passed"
        else:
            status = "failed"
        validation_log.update({account["name"]: {"derived": derived_value, "declared": declared_value, "status": status}})
        
        for i in account["items"]:
            inner_log = validate_account_totals(i)
            validation_log[account["name"]].update(inner_log)
            
    return validation_log
    
results = [validate_account_totals(j) for j in accounts.values()]
pprint(results)
```
